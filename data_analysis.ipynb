```python
from networkx.algorithms.community import greedy_modularity_communities
import matplotlib.pyplot as plt

# Load the dataset into a NetworkX graph object
G = nx.read_edgelist("/home/student/Desktop/222it029/Projects/WSC/WSC Project/Dataset/email-EuAll/Email-EuAll.txt")

# Compute the communities using the Louvain algorithm
communities = list(greedy_modularity_communities(G))

# Assign a different color to each community
colors = ["r", "b", "g", "c", "m", "y", "k"]
color_map = {}
for i, comm in enumerate(communities):
    for node in comm:
        color_map[node] = colors[i % len(colors)]

# Draw the graph with each community shown in a different color
nx.draw_networkx(G, node_color=[color_map[node] for node in G.nodes()], with_labels=False)
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Load the dataset into a NetworkX graph object
G = nx.read_edgelist("/home/student/Desktop/222it029/Projects/WSC/WSC Project/Dataset/email-EuAll/Email-EuAll.txt")

# Compute the degree centrality of each node
degree_centrality = nx.degree_centrality(G)

# Compute the betweenness centrality of each node
betweenness_centrality = nx.betweenness_centrality(G)

# Compute the eigenvector centrality of each node
eigenvector_centrality = nx.eigenvector_centrality(G)\

# Create a scatter plot of degree centrality vs. betweenness centrality
plt.scatter(degree_centrality.values(), betweenness_centrality.values())
plt.xlabel("Degree Centrality")
plt.ylabel("Betweenness Centrality")
plt.show()

# Create a scatter plot of degree centrality vs. eigenvector centrality
plt.scatter(degree_centrality.values(), eigenvector_centrality.values())
plt.xlabel("Degree Centrality")
plt.ylabel("Eigenvector Centrality")
plt.show()
```
